```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as c

Taste = c.Antecedent(np.arange(0, 11, 1), 'Taste')
Environment = c.Antecedent(np.arange(0, 11, 1), 'Environment')
Budget = c.Antecedent(np.arange(0, 80001, 1), 'Budget')
UserReview = c.Antecedent(np.arange(0, 6, 1), 'UserReview')
Preference = c.Consequent(np.arange(0, 10.1, 1), 'Preference')

Taste['low']=fuzz.gaussmf(Taste.universe, 0, 2)
Taste['medium']=fuzz.gaussmf(Taste.universe, 5, 3)
Taste['high']=fuzz.gaussmf(Taste.universe, 10, 2)

Environment['low']=fuzz.gaussmf(Environment.universe, 0, 2.5)
Environment['medium']=fuzz.gaussmf(Environment.universe, 6, 2.5)
Environment['high']=fuzz.gaussmf(Environment.universe, 10, 2.5)

Budget['low']=fuzz.gaussmf(Budget.universe, 0, 1500)
Budget['medium']=fuzz.gaussmf(Budget.universe, 3500, 1500)
Budget['high']=fuzz.gaussmf(Budget.universe, 8000, 1500)

UserReview['low']=fuzz.gaussmf(UserReview.universe, 0, 2)
UserReview['medium']=fuzz.gaussmf(UserReview.universe, 3, 2)
UserReview['high']=fuzz.gaussmf(UserReview.universe, 6 , 2)

Preference['Level1'] = fuzz.trimf(Preference.universe, [0, 2 ,4])
Preference['Level2'] = fuzz.trimf(Preference.universe, [2, 4, 6])
Preference['Level3'] = fuzz.trimf(Preference.universe, [5 , 6, 7])
Preference['Level4'] = fuzz.trimf(Preference.universe, [6 , 7, 8])
Preference['Level5'] = fuzz.trimf(Preference.universe, [8, 9, 10])

rule1 = c.Rule(Taste['low'] & Environment['low'] & Budget['low'] & UserReview['low'], Preference['Level1'])
rule2 = c.Rule(Taste['low'] & Environment['medium'] & Budget['low'] & UserReview['low'], Preference['Level1'])
rule3 = c.Rule(Taste['low'] & Environment['high'] & Budget['low'] & UserReview['low'], Preference['Level2'])
rule4 = c.Rule(Taste['low'] & Environment['low'] & Budget['medium'] & UserReview['low'], Preference['Level1'])
rule5 = c.Rule(Taste['low'] & Environment['medium'] & Budget['high'] & UserReview['low'], Preference['Level2'])
rule6 = c.Rule(Taste['low'] & Environment['low'] & Budget['low'] & UserReview['medium'], Preference['Level2'])
rule7 = c.Rule(Taste['low'] & Environment['medium'] & Budget['low'] & UserReview['high'], Preference['Level2'])
rule8 = c.Rule(Taste['medium'] & Environment['medium'] & Budget['low'] & UserReview['low'], Preference['Level2'])
rule9 = c.Rule(Taste['high'] & Environment['high'] & Budget['low'] & UserReview['low'], Preference['Level2'])
rule10 = c.Rule(Taste['high'] & Environment['high'] & Budget['high'] & UserReview['high'], Preference['Level4'])
rule11 = c.Rule(Taste['medium'] & Environment['medium'] & Budget['medium'] & UserReview['medium'], Preference['Level3'])
rule11 = c.Rule(Taste['medium'] & Environment['medium'] & Budget['high'] & UserReview['high'], Preference['Level4'])\

Preferencing_c = c.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7,rule8, rule9, rule10, rule11]) 

Preferencing = c.ControlSystemSimulation(Preferencing_c)

Preferencing.input['Taste'] = 9.0
Preferencing.input['Environment'] = 12.0
Preferencing.input['Budget'] = 13.0
Preferencing.input['UserReview'] = 5.0

Taste.view()
Environment.view()
Budget.view()
UserReview.view()


Preferencing.compute()
print(Preferencing.output['Preference'])
Preference.view(sim = Preferencing)

```
